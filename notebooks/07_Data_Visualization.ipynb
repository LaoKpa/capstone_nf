```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
f, ax = plt.subplots(figsize=(8, 6))

ax = sns.barplot(
    y= 'strat',
    x = 'rets', 
    orient='h',
    data = pd.DataFrame([['BEST MODEL', 18.42],['B&H', 12.48]], columns=['strat', 'rets']),
    #palette=("Blues_d")
    color='#DBF0FF',
)
plt.ylabel('Strategy', fontsize=18)
plt.xlabel('Total Returns (%)', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

```
